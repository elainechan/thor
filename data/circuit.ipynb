```python
# randomly chunk asset list to preset sizes
from itertools import islice
from random import randint

def random_chunk(li, min_chunk=4, max_chunk=12):
    it = iter(li)
    while True:
        nxt = list(islice(it,randint(min_chunk,max_chunk)))
        if nxt:
            yield nxt
        else:
            break
chunked_list = list(random_chunk(asset_list))
```
