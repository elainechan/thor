```python
# imports mta stops csv
stops = pd.read_csv('mtaStops.csv'); stops
```

```python
# iterate each row, place `lat` and `lon` column values into tuple, and insert in new column
latlon_list = []
for i, row in stops.iterrows():
    lat_str = str(stops.at[i, 'station_lat'])
    lon_str = str(stops.at[i, 'station_lon'])
    latlon_list.append((lat_str, lon_str))
latlon_series = pd.Series(latlon_list)
stops.insert(loc=4, column='station_lat_lon', value=latlon_series)
```

```python
# sets `station_lat_lon` column tuple figures to 6 decimal places (buggy)
for i in range(len(stops.index)-1):
    lat = tuple(stops.loc[i:i, ('station_lat_lon')])[0][0]
    lon = tuple(stops.loc[i:i, ('station_lat_lon')])[0][1]
    stops['station_lat_lon'][i] = (
        round(float(lat), 6),
        round(float(lon), 6)
    ) 
    
    # tuple(stops.loc[1:1, ('station_lat_lon')])
    # loc[:,('one','second')]
    # stops.loc[:, ('station_lat_lon', i, 0)]
    # stops.loc[:, ('station_lat_lon', i, 1)]
```
